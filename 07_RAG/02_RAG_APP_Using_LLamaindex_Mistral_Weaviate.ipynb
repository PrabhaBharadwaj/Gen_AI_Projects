# **Create Prompt template using  llama_index.core**

```python
'''
1. system(Set Behaviour of model) & assistant(LLM)
2. user'''
```

```python
from llama_index.core import PromptTemplate
```

```python
#Mentioning <|SYSTEM|> is default in llama_index to treat this as system prompt

system_prompt=""""
<|SYSTEM|># You are a Q&A assistant. Your goal is to answer questions as
accurately as possible based on the instructions and context provided
"""
```

```python
# This will wrap the default prompts that are internal to llama-index
query_wrapper_prompt = PromptTemplate("<|USER|>{query_str}<|ASSISTANT|>") # This means User Q, Model/Assistant o/p
```

# **Load the LLM Model**
- mistralai/Mistral-7B-Instruct-v0.1

- We are using huggingface_LLM from llama_index.llms.huggingface. this we can see in below Github. Each library or package will have its own github repo and inside base script we can see multiple classes and what are the parameter we need to pass for that class etc. Ex: class HuggingFaceLLM is there inside below link base.py

https://github.com/run-llama/llama_index/blob/main/llama-index-integrations/llms/llama-index-llms-huggingface/llama_index/llms/huggingface/base.py

```python
''' Class parameter detal
model_name: str = Field(
        default=DEFAULT_HUGGINGFACE_MODEL,
        description=(
            "The model name to use from HuggingFace. "
            "Unused if `model` is passed in directly."
        ),
    )
    context_window: int = Field(
        default=DEFAULT_CONTEXT_WINDOW,
        description="The maximum number of tokens available for input.",
        gt=0,
    )
    max_new_tokens: int = Field(
        default=DEFAULT_NUM_OUTPUTS,
        description="The maximum number of tokens to generate.",
        gt=0,
    )
    system_prompt: str = Field(
        default="",
        description=(
            "The system prompt, containing any extra instructions or context. "
            "The model card on HuggingFace should specify if this is needed."
        ),
    )
    query_wrapper_prompt: PromptTemplate = Field(
        default=PromptTemplate("{query_str}"),
        description=(
            "The query wrapper prompt, containing the query placeholder. "
            "The model card on HuggingFace should specify if this is needed. "
            "Should contain a `{query_str}` placeholder."
        ),
    )
    tokenizer_name: str = Field(
        default=DEFAULT_HUGGINGFACE_MODEL,
        description=(
            "The name of the tokenizer to use from HuggingFace. "
            "Unused if `tokenizer` is passed in directly."
        ),
    )
    device_map: str = Field(
        default="auto", description="The device_map to use. Defaults to 'auto'."
    )
    stopping_ids: List[int] = Field(
        default_factory=list,
        description=(
            "The stopping ids to use. "
            "Generation stops when these token IDs are predicted."
        ),
    )
    tokenizer_outputs_to_remove: list = Field(
        default_factory=list,
        description=(
            "The outputs to remove from the tokenizer. "
            "Sometimes huggingface tokenizers return extra inputs that cause errors."
        ),
    )
    tokenizer_kwargs: dict = Field(
        default_factory=dict, description="The kwargs to pass to the tokenizer."
    )
    model_kwargs: dict = Field(
        default_factory=dict,
        description="The kwargs to pass to the model during initialization.",
    )
    generate_kwargs: dict = Field(
        default_factory=dict,
        description="The kwargs to pass to the model during generation.",
    )
    is_chat_model: bool = Field(
        default=False,
        description=(
            LLMMetadata.__fields__["is_chat_model"].field_info.description
            + " Be sure to verify that you either pass an appropriate tokenizer "
            "that can convert prompts to properly formatted chat messages or a "
            "`messages_to_prompt` that does so."
        ),
    )
    '''
```

```python
from llama_index.llms.huggingface import HuggingFaceLLM
import torch #Use pytourch to load float16

#here parameter are defined inside base.py, same we copied and changed as per our llm model name
llm=HuggingFaceLLM(
    tokenizer_name="mistralai/Mistral-7B-Instruct-v0.1", #same model name used as tokenizer
    model_name="mistralai/Mistral-7B-Instruct-v0.1", # Mistral model name in hf
    context_window=4096, #I/P context window
    max_new_tokens=256, #O/P context window
    generate_kwargs={"temperature": 0.7, "do_sample": False},
    system_prompt=system_prompt,  #Set behaviour of model
    query_wrapper_prompt=query_wrapper_prompt, #user query and assistant/model response
    device_map="auto",
    stopping_ids=[50278, 50279, 50277, 1, 0],
    tokenizer_kwargs={"max_length": 4096},
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.float16}

)
```

# **Load Embedding model in LlamaIndex**

```python
%pip install llama-index-embeddings-huggingface -q
%pip install llama-index-embeddings-instructor -q
```

```python
# Import embedding model from hugging face
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
embed_model=HuggingFaceEmbedding(model_name="sentence-transformers/all-mpnet-base-v2")
```

# **VectorStore Index creation in LlamaIndex - Inmemory**
- This stores data in **in-memory**

```python
from llama_index.core import ServiceContext

#Github link  of service context:
# https://github.com/PrabhaBharadwaj/llama_index/blob/main/llama-index-core/llama_index/core/service_context.py

# ServiceContext: this is to set Chunk size etc parameter
Service_Context=ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm, #Mistral model
    embed_model=embed_model #define embedding model
)

#or Either this(Settings) way or we can use # ServiceContext: this is to set parameter
# All parameter which we set under Settings, will be passed automatically during index creation
#from llama_index.core import Settings
#Settings.chunk_size=512
#Settings.chunk_overlap=64
#Settings.llm=OpenAI(model="gpt-3.5-turbo")
#Settings.embed_model=OpenAIEmbedding(model="text-embedding-3-small")
```

```python
from llama_index.core import VectorStoreIndex

#Create vectore store index
index=VectorStoreIndex.from_documents(documents,  #Document
                                      service_context=Service_Context #embedding model and LLM Model
                                      )

  #index = MultiModalVectorStoreIndex.from_documents(documents,
  #                                                  service_context=Service_Context,  #this is to set Chunk size etc parameter
  #                                                  storage_context=storage_context, # This used to set External Vectordb
  #                                                  embed_model =embed_model)
```

## **Get final RAG response using LlamaIndex  Inmemory vectordb**

```python
#Here query_engine, Retrive from index/vectorsearch and passed to LLM(mentioned in service_context ) and provide result
query_engine=index.as_query_engine(k=1) #Retrives one top similarity search
```

```python
result=query_engine.query("what is Rag?")
```

## Create cluster in weaviate cloud page and use the API Key and WEAVIATE_CLUSTER below
- https://console.weaviate.cloud/signin?callbackUrl=https%3A%2F%2Fconsole.weaviate.cloud%2Fdashboard

```python
#Connect to OpenAI to get embedding model
from google.colab import userdata
WEAVIATE_CLUSTER=userdata.get('WEAVIATE_CLUSTER')
WEAVIATE_API_KEY=userdata.get('WEAVIATE_API_KEY')

import weaviate
#Connect to Weaviate cloud via API
weaviate_client=weaviate.Client(url=WEAVIATE_CLUSTER,auth_client_secret=weaviate.AuthApiKey(WEAVIATE_API_KEY))
```

```python
from llama_index.vector_stores.weaviate import WeaviateVectorStore
from llama_index.core import StorageContext
```

```python
# Define Weaviate client and its index name as Llamaindex
vector_store=WeaviateVectorStore(weaviate_client=weaviate_client,
                                 index_name="Llamaindex")

#define above one as parameter, here StorageContext is same like service context we used in inmemory db creation
storage_context = StorageContext.from_defaults(vector_store=vector_store)
```

```python
# Create llama index
index = VectorStoreIndex.from_documents(
                                  documents,
                                  storage_context=storage_context, #define weaviate cloud parameter
                                  embed_model=embed_model #embedding model
)
```

## **Get final RAG response using LlamaIndex- Weaviate cloud**

```python
#Here query_engine, Retrive from index/vectorsearch/weaviate cloud and passed to LLM and provide result
query_engine = index.as_query_engine(llm=llm)
```

```python
result3=query_engine.query("what is RAG evaluation method?")
```
