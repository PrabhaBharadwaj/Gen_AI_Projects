```python
#mport All the Required Libraries
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer
from transformers import AutoModelForCausalLM

import transformers
import torch  #Pytorch is better than Tensorflow/Keras, will have less steps
```

## **2. Define Llama2 Model**


### **2.1 Define Llama2 Model Name**
- We are using meta's official **llama2** model - **meta-llama/Llama-2-7b-chat-hf** via **Hugging face**
- if we dont have access to this official model, then we can use other person **daryl149/llama-2-7b-chat-hf**  model via **Hugging face**. Its same ll1m2 model only

```python
model="meta-llama/Llama-2-7b-chat-hf"
# model="daryl149/llama-2-7b-chat-hf"
```

### **2.2 Invoke Llam2 Model's Tokenizer**

- If we are using **hugging face pipeline**, then better to use specific model's tokenizer only, Directly doing langchain framework(Not hugging face pipeline) and it involves to write embedding in **vector DB**, That time we can use **Openai embedding model or hugging faces sentence-transformers**

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

### **2.3 Define Hugging face pipeline parameter**
- Here it downloads model

```python
pipeline=transformers.pipeline(
    "text-generation",    # Model name
    model=model,          # Llama2 Model
    tokenizer=tokenizer,  # Tokenizer
    torch_dtype=torch.bfloat16, # Default specification related to pytourch
    trust_remote_code=True,   # Default specification related to pytourch
    device_map="auto",    # Default specification related to pytourch
    max_length=1000,      # Fine tuned new model O/P length
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id # Default specification related to pytourch
    )
```

### **2.4 Invoke Llama2 model via Hugging face pipeline**
- Execute LLM Wrapper

```python
llm=HuggingFacePipeline(pipeline=pipeline, model_kwargs={'temperature':0})
```

## **3. Pass the prompt to get text Generation**

```python
prompt="What would be a good name for a company that makes colorful socks"
print(llm(prompt))
```

```python
prompt="I want to open a restaurant for  indian food. Suggest me a fence name for this"
print(llm(prompt))
```
