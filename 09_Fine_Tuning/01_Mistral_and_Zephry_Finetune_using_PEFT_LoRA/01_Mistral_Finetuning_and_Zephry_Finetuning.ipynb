## **Fine Tuning**
- Foundational model some time wont be good enough to our task, so we may need to make it good to understand our data and provide task specific response.
- For that we can do by below 3 ways:
  - **1. Prompt engineering**
  - **2. Fine Tuning**
  - **3. RAG**

- In **Fine Tuning** we are adjusting the original parameter/weight value by training using new data

- There are Multiple type of **Fine tuning**,

  - **1. Supervised/Instruction fine tuning**
    - It will have dependentet and independent column with labeled dat during training
    - Retraining **Entire/Full** weight here
    - So itss too costly/Computational expensive and need super computer/GPU and long time
  - **2. PEFT - Parameter Efficient Fine Tuning/ Targeted Fine tuning**
    - Here only last layer or few layers parameter/weights only retrained
    - It updates only small set of parameter.
    - Few PEFT methods are LoRa and QLoRA
  - **3. RLHF - ReiNforcement Learning Through Human Feedback/Instruction fine tuning**
    - Here also we adjust the weighting
  - **4. Quanization:**
    - 2 type Quaantization
        - **NF4 Quantization**
        - **Double Quantization**
    - Here we will minimize the original parameter/weight value.
    - Ex: If its 1.572016 , then trim to integer 1.57 or 1
    - Ex: **64bit, 16bit, 8bit,4bit Quantizations** Model. Here we converts models parameter to these bit value

- **5. DPO - Direct Performance Optimization:**

### **Download Samsum data from hugging face**
- The SAMSum dataset contains about 16k messenger-like conversations with summaries. Conversations were created and written down by linguists fluent in English.

```python
df=load_dataset("samsum",split="train")
print(df.shape)
```

```python
datadf=df.to_pandas()
datadf.head()
```

```python
datadf=datadf.sample(50)
# Create new column "text" by combining 2 column
datadf["text"]=datadf[["dialogue", "summary"]].apply(lambda x: "###Human: Summarize this following dialogue: " + x["dialogue"] + "\n###Assistant: " +x["summary"], axis=1 )
```

## **Import Tokenizer from Mistral model using Transformer(AutoTokenizer)**
- This is to tokenize our data. Here we used **Mistral's - Mistral-7B-Instruct-v0.1-GPTQ** Tokenizer
- https://huggingface.co/TheBloke/Mistral-7B-Instruct-v0.1-GPTQ



```python
from transformers import AutoModelForCausalLM, AutoTokenizer, GPTQConfig, TrainingArguments
```

```python
#Load this mistral's base tokenizer model from HF
tokenizer=AutoTokenizer.from_pretrained("TheBloke/Mistral-7B-Instruct-v0.1-GPTQ")
```

```python
#End of tokenizer is </s>
tokenizer.eos_token
```

```python
tokenizer.pad_token=tokenizer.eos_token
```

### **GPTQConfig**
- Configure with GPTQConfig quantization .
- This is wrapper class for quantization. Quantization techniques reduce memory and computational costs by representing weights and activations with lower-precision data types like 8-bit integers (int8).

```python
# Using GPTQConfig - quantization method and setting quantization bits=4
quantization_config_loading = GPTQConfig(bits=4, disable_exllama=True, tokenizer=tokenizer)
```

## **Import Model from Mistral  using Transformer(AutoModelForCausalLM)**
- Defined quantization_config frm **GPTQConfig**. This way we download 4 bit quantized mistral model

```python
#Load this mistral's base Model model from HF
model=AutoModelForCausalLM.from_pretrained( "TheBloke/Mistral-7B-Instruct-v0.1-GPTQ",
                                     quantization_config=quantization_config_loading,
                                      device_map="auto")
```

```python
print(model)
```

### **Set some configuration in the above downloaded model**

```python
model.config.use_cache=False
model.config.pretraining_tp=1
model.gradient_checkpointing_enable()
```

## **Fine Tune the Mistral model**

```python
# prepare_model_for_kbit_training is a library from peft pkg for making model to ready for finetuning
from peft import prepare_model_for_kbit_training
```

```python
#prepare_model_for_kbit_training is a library from peft pkg for making model to ready for finetuning

model = prepare_model_for_kbit_training(model)
```

### **Set LoraConfig**

```python
from peft import LoraConfig
```

```python
peft_config=LoraConfig(
    r=16,
    lora_alpha=16,
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules= ["q_proj", "v_proj"]
)
```

### **Create final fine tuned model object by passing peft config**

```python
from peft import get_peft_model
```

```python
#Create final fine tuned model by passing peft config
model=get_peft_model(model,peft_config)
```

### **Train the peft model using SFTTrainer passing TrainingArguments**

```python
from trl import SFTTrainer
from transformers import TrainingArguments  # This is to set training argument
```

```python
#Set Training argument
training_arguments = TrainingArguments(
        output_dir="mistral-finetuned-samsum",  #Fine tuned new model name
        per_device_train_batch_size=8,
        gradient_accumulation_steps=1,
        optim="paged_adamw_32bit",
        learning_rate=2e-4,
        lr_scheduler_type="cosine",
        save_strategy="epoch",
        logging_steps=100,
        num_train_epochs=1,
        max_steps=250,
        fp16=True,
        push_to_hub=True
    )
```

```python
notebook_login()
```

```python
#Use SFTTrainer to Set the final training object with model name, token , peft config etc
trainer=SFTTrainer(
    model=model,  # To be fine tuned Mistral model name
    train_dataset=data,  # Training dataset
    peft_config=peft_config,  #Peft config
    dataset_text_field="text",  #I/P Field which used for training
    args=training_arguments,  # Training parameter
    tokenizer=tokenizer,   #Tokenizer
    packing=False,
    max_seq_length=512

)
```

### Import Libraries

```python
import torch
from datasets import load_dataset, Dataset
from peft import LoraConfig, prepare_model_for_kbit_training, get_peft_model
from transformers import AutoModelForCausalLM, AutoTokenizer, GPTQConfig, TrainingArguments
from trl import SFTTrainer
```

## List all the Fine tuning model param/config

```python
class Config:
    MODEL_ID = "TheBloke/zephyr-7B-alpha-GPTQ"  # Going to Finetune model
    DATASET_ID = "bitext/Bitext-customer-support-llm-chatbot-training-dataset"  #Dataset which going to use for FT
    CONTEXT_FIELD= ""
    INSTRUCTION_FIELD = "instruction"  #I/P feature
    TARGET_FIELD = "response"  #Target col
    BITS = 4
    DISABLE_EXLLAMA = True
    DEVICE_MAP = "auto"
    USE_CACHE = False
    LORA_R = 16
    LORA_ALPHA = 16
    LORA_DROPOUT = 0.05
    BIAS = "none"
    TARGET_MODULES = ["q_proj", "v_proj"]
    TASK_TYPE = "CAUSAL_LM"
    OUTPUT_DIR = "zephyr-support-chatbot"
    BATCH_SIZE = 8
    GRAD_ACCUMULATION_STEPS = 1
    OPTIMIZER = "paged_adamw_32bit"
    LR = 2e-4
    LR_SCHEDULER = "cosine"
    LOGGING_STEPS = 50
    SAVE_STRATEGY = "epoch"
    NUM_TRAIN_EPOCHS = 1
    MAX_STEPS = 250
    FP16 = True
    PUSH_TO_HUB = True
    DATASET_TEXT_FIELD = "text"
    MAX_SEQ_LENGTH = 512
    PACKING = False


```

## Main class to Fine tune the model

```python
class ZephyrTrainer:
  def __init__(self):
    self.config = Config()  # Pass config list from Config Class
    self.tokenizer = AutoTokenizer.from_pretrained(self.config.MODEL_ID)  #Invoke Tokenizer: Pass MODEL_ID ("TheBloke/zephyr-7B-alpha-GPTQ"'S ) tokenizer
    self.tokenizer.pad_token = self.tokenizer.eos_token

  # Write system prompt used to FT model. Here we pass few shot sample also with User Q and Target Q
  def process_data_sample(self,example):
        processed_example = """<|system|>\n You are a support chatbot who helps with user queries chatbot
        who always responds in the style of a professional.\
        \n<|user|>\n"""  + example[self.config.INSTRUCTION_FIELD]  + "\n<|assistant|>\n" + example[self.config.TARGET_FIELD]

        return processed_example

  #Dataset creation function
  def create_dataset(self):
        data = load_dataset(self.config.DATASET_ID, split="train") #split the data

        print("\n====================================================================\n")
        print("\t\t\tDOWNLOADED DATASET")
        print("\n====================================================================\n")

        #Take only INSTRUCTION_FIELD=instruction and TARGET_FIELD=response
        df = data.to_pandas()
        df[self.config.DATASET_TEXT_FIELD] = df[[self.config.INSTRUCTION_FIELD, self.config.TARGET_FIELD]].apply(lambda x: self.process_data_sample(x), axis=1)

        print("\n====================================================================\n")
        print("\t\t\tPROCESSED DATASET")
        print(df.iloc[0])
        print("\n====================================================================\n")

        processed_data = Dataset.from_pandas(df[[self.config.DATASET_TEXT_FIELD]])
        return processed_data



  def prepare_model(self):

        #Quantization config
        ### **GPTQConfig**
        ## Configure with GPTQConfig quantization .
        # This is wrapper class for quantization. Quantization techniques reduce memory and computational costs by
        # representing weights and activations with lower-precision data types like 8-bit integers (int8).

        bnb_config = GPTQConfig(
                                    bits=self.config.BITS,  #BITS =4
                                    disable_exllama=self.config.DISABLE_EXLLAMA,
                                    tokenizer=self.tokenizer
                                )

        #Load the model from HF
        model = AutoModelForCausalLM.from_pretrained(
                                                        self.config.MODEL_ID,
                                                        quantization_config=bnb_config,
                                                        device_map=self.config.DEVICE_MAP
                                                    )

        print("\n====================================================================\n")
        print("\t\t\tDOWNLOADED MODEL")
        print(model)
        print("\n====================================================================\n")

        model.config.use_cache=self.config.USE_CACHE
        model.config.pretraining_tp=1
        model.gradient_checkpointing_enable()
        model = prepare_model_for_kbit_training(model)

        print("\n====================================================================\n")
        print("\t\t\tMODEL CONFIG UPDATED")
        print("\n====================================================================\n")

        #LORA Configuration
        peft_config = LoraConfig(
                                    r=self.config.LORA_R,
                                    lora_alpha=self.config.LORA_ALPHA,
                                    lora_dropout=self.config.LORA_DROPOUT,
                                    bias=self.config.BIAS,
                                    task_type=self.config.TASK_TYPE,
                                    target_modules=self.config.TARGET_MODULES
                                )

        model = get_peft_model(model, peft_config)

        print("\n====================================================================\n")
        print("\t\t\tPREPARED MODEL FOR FINETUNING")
        print(model)
        print("\n====================================================================\n")

        return model, peft_config

  # Set training argument
  def set_training_arguments(self):
        training_arguments = TrainingArguments(
                                                output_dir=self.config.OUTPUT_DIR,
                                                per_device_train_batch_size=self.config.BATCH_SIZE,
                                                gradient_accumulation_steps=self.config.GRAD_ACCUMULATION_STEPS,
                                                optim=self.config.OPTIMIZER,
                                                learning_rate=self.config.LR,
                                                lr_scheduler_type=self.config.LR_SCHEDULER,
                                                save_strategy=self.config.SAVE_STRATEGY,
                                                logging_steps=self.config.LOGGING_STEPS,
                                                num_train_epochs=self.config.NUM_TRAIN_EPOCHS,
                                                max_steps=self.config.MAX_STEPS,
                                                fp16=self.config.FP16,
                                                push_to_hub=self.config.PUSH_TO_HUB
                                            )

        return training_arguments


  #Train the model
  def train(self):
    data=self.create_dataset()  # Craete dataset
    model,peft_config=self.prepare_model()  #Prepare model and peft config
    training_args=self.set_training_arguments()  #Set training argument

    #Use SFTTrainer to Set the final training object with model name, token , peft config etc
    trainer=SFTTrainer(
        model=model,  # To be fine tuned Zephry model name
        train_dataset=data,  # Training dataset
        peft_config=peft_config,  #Peft config
        dataset_text_field=self.config.DATASET_TEXT_FIELD,  #I/P Field which used for training
        args=training_args,  # Training parameter
        tokenizer=self.tokenizer,  #Tokenizer
        packing=self.config.PACKING,
        max_seq_length=self.config.MAX_SEQ_LENGTH

    )

    #Train the model
    trainer.train()
    print("\n====================================================================\n")
    print("\t\t\tFINETUNING COMPLETED")
    print("\n====================================================================\n")

    #Push the model to HF
    trainer.push_to_hub()
```

```python
#Create ZephyrTrainer object
zephyr_trainer = ZephyrTrainer()
```

### Execute Fine tune train method

```python
# Execute train function inside class to train the model
zephyr_trainer.train()
```
